```python
# Dependencies and setup
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
from sklearn.linear_model import Ridge
import seaborn as sns
```

```python
# Convert csv file to dataframe
annual_factors = Path("Resources/Annual_Macroeconomic_Factors.csv")
annual_factors_pd = pd.read_csv(annual_factors)

# Set Date as index, avoids having to plot xticks
annual_factors_pd = annual_factors_pd.set_index("Date")
annual_factors_pd
```

```python
# Narrowing data specifically to look at the House Price Index 
house_index = annual_factors_pd["House_Price_Index"]

# Make bar graph of HPI vs Year
bar_graph = house_index.plot(kind="bar", figsize=(12,6))
bar_graph.set_title("HPI by Year")
bar_graph.set_ylabel("House Price Index")
bar_graph.set_xlabel("Year (In January)")
bar_graph
```

```python
# Repeat the above cell with unemployment rate instead of HPI
unemployment = annual_factors_pd["Unemployment_Rate"]

# Make another bar graph for unemployment vs year
bar_graph = unemployment.plot(kind="bar", figsize=(12, 6))
bar_graph.set_title("Unemployment Rate by Year")
bar_graph.set_ylabel("Unemployment Rate (%)")
bar_graph.set_xlabel("Year (In January)")
bar_graph
```

```python
# repeat above cells for GDP
gdp = annual_factors_pd["Real_GDP"]

#make bar graph showing GDP vs time
bar_graph = gdp.plot(kind="bar", figsize=(12, 6))
bar_graph.set_title("GDP Growth Over Time")
bar_graph.set_ylabel("GDP")
bar_graph.set_ylabel("Year (In January)")
bar_graph
```

```python
# repeat above cells for CPI
cpi = annual_factors_pd["Consumer_Price_Index"]

# creat bar graph comparing CPI over time
bar_graph = cpi.plot(kind="bar", figsize=(12, 6))
bar_graph.set_title("CPI by Year")
bar_graph.set_ylabel("CPI")
bar_graph.set_xlabel("Year (In January)")
bar_graph
```

```python
# scatter plot comparing gdp over time to hpi over time
plt.scatter(cpi, house_index, marker="o", facecolors="red", edgecolors="black", alpha=0.75, )
plt.title("CPI vs HPI")
plt.xlabel("CPI Over Time")
plt.ylabel("HPI Over Time")

# add correlation coefficient (r)
r, p = stats.pearsonr(cpi, house_index)
plt.annotate("r = {:.2f}".format(r), xy=(0.7, 0.9), xycoords="axes fraction")
```

- The first four figures are bar graphs showing the progression and change in each macroeconomic factor from January 1975 to January 2021
- The fifth figure shows the relationship and correlation coefficient between CPI and House Price Index over the same period of time
    - As is noticeable on the plot, CPI and HPI have a strong relationship/correlation with one another
- The bar graphs created for GDP, CPI, and House Price Index show a steady overall increase over that timeframe
- The bar graph for unemployment rate shows a fluctuation of increased and decreased unemployment rate, something that is to be expected given the volatility of the economy
    - It is worth noting that given the steady overall increase in GDP, CPI, and House Price Index, one would expect unemployment to steadily decrease
- While the csv file did include real disposable income as a factor, household income is usually a better indicator of purchasing power when it comes to housing purchases
    - Because this csv did not contain that information, it was obtained from the US Census Bureau's website and will be reflected in the analysis page and presentation
