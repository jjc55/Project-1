```python
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path

import requests
import time
from scipy.stats import linregress
```

```python
annual_factors = Path("Resources/Annual_Macroeconomic_Factors.csv")
annual_factors_pd = pd.read_csv(annual_factors)
annual_factors_pd = annual_factors_pd.set_index("Date")
annual_factors_pd.head()
```

```python
#build a scatter plot for Real Disposable Income & Unemployment
plt.scatter(annual_factors_pd ["Real_Disposable_Income"], annual_factors_pd ["Unemployment_Rate"])
plt.ylabel("Unemployment")
plt.xlabel("Real Disposable Income")
plt.grid(True)
plt.show()
```

```python
#Line graph of Real Disposable Income
real_disposable_income = annual_factors_pd["Real_Disposable_Income"]
line_graph_income = real_disposable_income.plot(kind="line", figsize=(12,6))
line_graph_income.set_title("Income")
line_graph_income.set_ylabel("Real Disposable Income")
line_graph_income.set_xlabel("Date")
line_graph_income
```

```python
#Line graph of Unemployment Rate
unemployment_rate = annual_factors_pd["Unemployment_Rate"]
line_graph_unemployment = unemployment_rate.plot(kind="line", figsize=(12,6))
line_graph_unemployment.set_title("Unemployment")
line_graph_unemployment.set_ylabel("Unemployment Rate")
line_graph_unemployment.set_xlabel("Date")
line_graph_unemployment
```

```python
# Define a function to create Linear Regression plots
def linear_regression_plots_function (x_values, y_values, text_coordinates, label_x, label_y):
    (slope, intercept, rvalue, pvalue, stderr) = linregress (x_values,y_values)
    string = f"y={round(slope,2)}x + {round(intercept,2)}"
    linear_regression = slope * x_values + intercept

    print (f"r-value = {rvalue}")
    plt.scatter(x_values, y_values)
    plt.plot(x_values,linear_regression, "r-")
    plt.annotate(string, text_coordinates, color = "red", fontsize = 12)
    plt.xlabel(label_x)
    plt.ylabel(label_y)
    plt.show()
```

```python
# Linear regression
x_values = annual_factors_pd ["Real_Disposable_Income"]
y_values = annual_factors_pd ["Unemployment_Rate"]
linear_regression_plots_function(x_values, y_values, (25000,9), "Real Disposable Income", "Unemployment Rate")


```

-Figure 1 is a scatterplot that evaluates the relationship between the macroeconomic factors
unemployment rate and real disposable income
-The data set included yearly data from 1975 to 2021
-Results show low negative correlation
-Figures 2 and 3 show these macroeconomic factors individually
line graphs of unemployment over time and income over time
While income steadily increases, the unemployment rate is variable over time
-Conclusion: a r-value of -0.34 shows a low negative correlation. When income is low, unemployment is high. When income is high, unemployment is low. The correlation is weaker than hypothesized which may be due to other macroeconomic factors. Further exploration into other macroeconomic factors is recommended for future research.
